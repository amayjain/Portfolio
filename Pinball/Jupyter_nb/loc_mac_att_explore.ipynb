```python
# 16 rows have 0 machines
# empty [] in location_machine_xrefs
df['location_machine_xrefs'].value_counts()[[0]]
```

```python
delete_indices = list(df[df['location_machine_xrefs'] == '[]'].index)
delete_indices
```

```python
df = df.drop(delete_indices)
```

```python
df.shape
```

```python
df.index = range(len(df))
```

```python
machines = df['location_machine_xrefs'].copy()
```

```python
df.drop(['location_machine_xrefs'], axis = 1, inplace= True)
```

```python
machines = machines.apply(eval)
```

```python
baseline = pd.json_normalize(machines[0])
baseline
```

```python
store = []
store.append(baseline)
```

```python
for i in range(1, len(machines)):
    store.append(pd.json_normalize(machines[i]))
```

```python
all_machines = pd.concat(store)
```

```python
all_machines
```

```python
all_machines.index = range(len(all_machines))
```

```python
all_machines.columns = ['id', 'created_at', 'updated_at', 'location_id', 'machine_id',
       'ic_enabled', 'machine_id', 'machine_name', 'machine_updated_at',
       'machine_ipdb_link', 'machine_year', 'machine_manufacturer',
       'machine_machine_group_id', 'machine_ipdb_id', 'machine_opdb_id']
```
