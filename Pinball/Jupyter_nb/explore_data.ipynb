```python
# find out inconsistencies in images column
index_fix = []
for i in range(1, len(data_copy)):
    if (len(data_copy['images'][i]) > 1):
        index_fix.append(i)
```

```python
index_fix
```

```python
names = []
for i in index_fix:
    names.append(data_copy.iloc[i]['name'])
```

```python
names
```

```python
for i in index_fix:
    data_copy['images'][i] = data_copy['images'][i][0]
```

```python
newdf = pd.json_normalize(data_copy['images'][0])
print(newdf.shape)
for i in range(1, len(data_copy)):
    convert_df = pd.json_normalize(data_copy['images'][i])
    newdf = pd.concat([newdf, convert_df])   
```

```python
new_index = pd.Index(range(len(data_copy)))
newdf = newdf.set_index(new_index)
newdf
```

```python
df1 = pd.concat([data_copy,newdf], axis = 1)
df1.shape
```

```python
df1 = df1.drop('images', axis = 1)
df1.shape
```

```python
df1.columns
```

```python
format_rows = list(df1[df1['title'] == "-1"][newdf.columns].index)
```

```python
df2 = df1.loc[format_rows].copy()
```

```python
df1 = df1.drop(format_rows)
```

```python
df1 = df1.set_index(pd.Index(range(len(df1))))
df1
```

```python
df2 = df2.set_index(pd.Index(range(len(df1), len(df1) + len(df2))))
df2
```

```python
df2_1 = df2[newdf.columns].copy()
```

```python
df2 = df2.drop(newdf.columns, axis = 1)
```

```python
df2_1[:] = np.nan
```

```python
df2 = pd.concat([df2, df2_1], axis = 1)
```

```python
clean_data = pd.concat([df1,df2])
clean_data
```

```python
clean_data['manufacture_date'] = pd.to_datetime(clean_data['manufacture_date'])
```

```python
clean_data['keywords'] = clean_data['keywords'].apply(eval)
clean_data['features'] = clean_data['features'].apply(eval)
```

```python
def join(x):
    if len(x) > 0:
        word = ""
        for i in range(len(x)):
            word = word + x[i] + " "
        return word
    else:
        return np.nan
```

```python
clean_data['keywords'] = clean_data['keywords'].apply(join)
clean_data['features'] = clean_data['features'].apply(join)
```
